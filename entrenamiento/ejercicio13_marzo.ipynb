```python
import pandas as pd
import math
# librerías

#variables
m=0.849
b=1.509
# datos
datos1 = [1,2,2,3,4,4,5,6]
datos2 = [2,3,4,4,4,6,5,7]

# diccionario
diccionario = {"x":datos1,
               "y":datos2}

# llamar pandas
df = pd.DataFrame(diccionario)

# nuevas columnas
df['ypred'] = m* df['x'] + b

df['y-pred'] = df['y'] - df['ypred']

df['y-pred^2'] = df['y-pred'] ** 2


df
# fila con sumas
#df.loc['Total'] = df.sum()

# mostrar dataframe
df

```

```python
total = df['y-pred^2'].sum()
total

cantidad = len(df)
cantidad

promedio = total / cantidad
promedio

raiz = math.sqrt(promedio)
raiz
```

```python
import numpy as np
import matplotlib.pyplot as plt

# grafica de la informacion
plt.scatter(df['x'], df['y'], color='red', label='y')
plt.scatter(df['x'], df['ypred'], color='blue', label='ypred')

# línea de regresión

x_valor = np.linspace(0,7)
y_valor = m*x_valor + b

plt.plot(x_valor, y_valor, color='green', label='Regresión Lineal')

# etiquetas y estilo
plt.xlabel('valor de x')
plt.ylabel('valor de y')
plt.title('Recta de Regresión Lineal ejemplo excel')
plt.legend()
plt.grid()



```
