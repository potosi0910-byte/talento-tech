```python

# librerías
import pandas as pd

# datos
datos1 = [1,2,2,3,4,4,5,6]
datos2 = [2,3,4,4,4,6,5,7]

# diccionario
diccionario = {"x":datos1,
               "y":datos2}

# llamar pandas
df = pd.DataFrame(diccionario)

# nuevas columnas
df['xy'] = df['x'] * df['y']
df['x2'] = df['x'] ** 2

df
# fila con sumas
df.loc['Total'] = df.sum()

# mostrar dataframe
print(df)
df

```

```python
""" hasta aqui ps la tabla con sus sumas """

# n = número de observaciones
n = len(datos1)
# suma de x
sum_x = df.loc['Total','x']

# suma de y
sum_y = df.loc['Total','y']

# suma de xy
sum_xy = df.loc['Total','xy']

# suma de x²
sum_x2 = df.loc['Total','x2']

""" aqui ps los resultados de la tarea segun la formula"""
# calcular pendiente (b)
b = (n*sum_xy - (sum_x*sum_y)) / (n*sum_x2 - (sum_x**2))
print(b)

# calcular intercepto (a)
a = (sum_y/n) - b*(sum_x/n)

# mostrar resultados
print("Pendiente (b):", b)
print("Intercepto (a):", a)

# ecuación de la recta de regresión
print("Ecuación de regresión:")
print(f"y = {a} + {b}x")

pd.DataFrame(df)
```

```python
# grafica donde se importan las bibliotecas para la regresion lineal

import matplotlib.pyplot as plt
import numpy as np

# puntos originales
plt.scatter(df['x'], df['y'], color='red', label='Datos originales', zorder=5)

# línea de regresión
x_vals = np.linspace(df['x'].min(), df['x'].max(), 100)
y_vals = a + b * x_vals
plt.plot(x_vals, y_vals, color='red', label=f'y = {a:.2f} + {b:.2f}x')

# etiquetas y estilo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Recta de Regresión Lineal')
plt.legend()
plt.grid(True)
plt.show()
```
