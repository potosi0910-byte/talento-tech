```python
import pandas as pd
import math
# librerías


# datos
datos1 = [78.5,74.3,104.3,87.6,95.9,109.2,102.7,72.5,93.1,115.9,83.8,113.3,109.4]
datos2 = [7,1,11,11,7,11,3,1,2,21,1,11,10]
datos3 = [26,29,56,31,52,55,71,31,54,47,40,66,68]
datos4 = [6,15,8,8,6,9,17,22,18,4,23,9,8]
datos5 = [60,52,20,47,33,22,6,44,22,26,34,12,12]

# diccionario
diccionario = {"y":datos1,
               "x1":datos2,
               "x2":datos3,
               "x3":datos4,
               "x4":datos5,}

#tabla
df= pd.DataFrame(diccionario)
df

```

```python
# Replace the 'y' column with ones
df_modified = df.copy()
df_modified['y'] = 1

# Convert the modified DataFrame to a NumPy matrix
matrix_with_y_as_ones = df_modified.values

# Display the resulting matrix
print(matrix_with_y_as_ones)
```

```python
# Transpose the matrix
transposed_matrix = matrix_with_y_as_ones.T

# Display the transposed matrix
print(transposed_matrix)
```

```python
import numpy as np

# Extract the 'y' column as a NumPy array
y_column_vector = df['y'].values.reshape(-1, 1)

# Display the column vector
print(y_column_vector)
```

```python
import numpy as np

# Multiply the transposed matrix by the original matrix
result_matrix = np.dot(transposed_matrix, matrix_with_y_as_ones)

# Display the resulting matrix
print(result_matrix)
```

```python
import numpy as np

# Calculate the inverse of the result_matrix
inverse_result_matrix = np.linalg.inv(result_matrix)

# Display the inverse matrix
print(inverse_result_matrix)
```

```python
import numpy as np

# Multiply inverse_result_matrix by transposed_matrix
step1_multiplication = np.dot(inverse_result_matrix, transposed_matrix)

# Multiply the result by y_column_vector
final_result_beta = np.dot(step1_multiplication, y_column_vector)

# Display the final result (which should be the beta coefficients)
print(final_result_beta)
```
